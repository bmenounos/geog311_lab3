## Directional Data: 20 points

Analysis of directional data often reveals information about the type or nature of geomorphic processes. For example, large-scale lithologic structures (e.g. anticline) and bedding planes may explain the spatial distribution of rock avalanches. It is therefore important to investigate whether landslides show a preferred orientation. To work with directional data, however, one can not simply take the 'average' of circular data expressed in degrees (e.g. consider the average direction of the compass bearings 350 $^{\circ}$,  355 $^{\circ}$, and 23$^{\circ}$ ). To determine statistical parameters such as central tendency and dispersion, we must convert the observations to vectors and plot them on a unit circle. A compass direction can be converted into polar coordinates by the following:

\begin{equation}
\begin{split}
x_{i} &= cos(\theta_{i}) \\
y_{i} &= sin(\theta_{i}) \\
\end{split}
\end{equation}

Where $x_{i}$ and $y_{i}$ represent the X and Y coordinates of the unit vector and $\theta_{i}$ is the compass direction (in radians).

Mean direction ($\bar{\theta}$) for a set of 2-D vectors can be obtained by first computing the vector resultant:

\begin{equation}
\begin{split}
X_{r} &= \sum^{n}_{i=1} cos(\theta_{i}) \\ 
Y_{r} &= \sum^{n}_{i=1} sin(\theta_{i}) \\
\end{split}
\end{equation}

Where $X_{r}$ and $Y_{r}$ are the X and Y coordinates of the resultant.



$\bar{\theta}$ is then obtained from:

\begin{equation}
\bar{\theta}=tan^{-1}(\frac{Y_r}{X_r})
\end{equation}

The length ($R$) of the resultant (a measure of vector dispersion) can be found by:
\begin{equation}
R=\sqrt{X_{r}^{2}+Y_{r}^{2}} = \sqrt{(\sum_{i=1}^{n}cos \theta_{i})^2+(\sum_{i=1}^{n} sin \theta_{i})^2}
\end{equation}


In order to compare resultants from different sample sizes the standardized resultant ($\bar{R}$) is usually calculated:

\begin{equation}
\bar{R} = \frac{R}{n} 
\end{equation}

where n is the number of observations. Finally the circular variance ($s^{2}_{o}$) can be obtained from:

\begin{equation}
s^{2}_{o} = 1 - \bar{R}
\end{equation}
